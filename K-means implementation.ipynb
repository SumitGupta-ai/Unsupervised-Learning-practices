```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_clustering.csv')
df
```

```python
df.info()
```

```python
plt.scatter(x='cgpa', y='iq', data=df)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wssc = []
for cent in range(1,10):
    km = KMeans(n_clusters=cent)
    km.fit_predict(df)
    wssc.append(km.inertia_)
    
```

```python
plt.plot(range(1,10),wssc)
plt.show()
```

```python
x = df.iloc[:,:].values
kme = KMeans(n_clusters=4)
prd = kme.fit_predict(x)
```
